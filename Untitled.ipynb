# 2. Find out the names and prices of all the apps that have prices more than 1000.


```python
df['Price'].value_counts()
```

```python
df['Price_real'] = np.where(df['Price']=='Free', -1, df.Price)
```

```python
df.head()
```

```python
df1 = df[df['Price_real'].notnull()]
df1.isnull().sum()
```

```python
df1['Price_real'] = df1['Price_real'].astype(str).apply(lambda x: x.replace('₹ ',''))
df1['Price_real'] = df1['Price_real'].astype(str).apply(lambda x: x.replace(',',''))
df1['Price_real'] = df1['Price_real'].astype(float)
```

```python
df1['Price_real'].value_counts()
```

```python
df2 = df1[df1['Price_real']>1000.0]
```

```python
df2.groupby(["Name"]).sum().sort_values("Price_real")
```

In the above code, we finally got our answer grouped by Name.

# 5. Divide the ratings into three categories: poor(rating less than 2.5 ), average(rating between 2.5 and 3.5) and good(rating greater than 3.5 ) and find out the count for each rating.


```python
df['Rating'].unique()
```

```python
df.info()
```

```python
#function to change the value of rating column
rating = df['Rating']
def rating(row):
    if (row['Rating'] < 2.5):
        return 'Poor'
    elif (row['Rating'] > 3.5):
        return 'Good'
    else:
        return 'Average'          
```

```python
df['Rating']=df.apply(lambda row: rating(row),axis=1) 
```

```python
df['Rating']
```

```python
df['Rating'].value_counts()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Rating', data=df)
```
